```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def print_graphs(list_graphs, xticks=[1,2]):
  bar_width = 0.2

  # Gerar o gráfico
  plt.figure(figsize=(10, 6))

  n = len(list_graphs)
  for i, graph in enumerate(list_graphs):
    offset = (i - n / 2) * bar_width
    plt.bar(graph['data']['hops'] + offset, graph['data']['_Eihop'], bar_width, label=graph['label'], color=graph['color'])


  plt.title('Gráfico dos Dados')
  plt.xlabel('hops')
  plt.ylabel('_Eihop')

  # Definir os ticks do eixo x
  plt.xticks(xticks)

  # Adicionar legenda
  plt.legend()

  # Mostrar o gráfico
  plt.show()
```

```python
#Desvio padrão no gráfico
# Calcular o desvio padrão para cada conjunto de dados
list_graphs = [
  {
    'label': 'CSMA',
    'data': csma_df
  },
  {
    'label': 'ALOHA sem RDC',
    'data': aloha_radio_always_on_df
  },
  {
    'label': 'ALOHA RDC 5% Duty Cycle',
    'data': aloha_rdc_5_duty_cycle
  },
  {
    'label': 'ALOHA RDC 10% Duty Cycle',
    'data': aloha_rdc_10_duty_cycle
  },
  {
    'label': 'ALOHA RDC 20% Duty Cycle',
    'data': aloha_rdc_20_duty_cycle
  },
  {
    'label': 'ALOHA RDC 30% Duty Cycle',
    'data': aloha_rdc_30_duty_cycle
  },
  {
    'label': 'ALOHA RDC 40% Duty Cycle',
    'data': aloha_rdc_40_duty_cycle
  },
  {
    'label': 'ALOHA RDC 50% Duty Cycle',
    'data': aloha_rdc_50_duty_cycle
  },
  {
    'label': 'ALOHA RDC 60% Duty Cycle',
    'data': aloha_rdc_60_duty_cycle
  }

]

std_devs = [df['data']['_Eihop'].std() for df in list_graphs]

# Criar uma lista com os rótulos para cada conjunto de dados
labels = [df['label'] for df in list_graphs]

# Criar uma lista com as posições no eixo x para cada barra
x_pos = np.arange(len(labels))

# Plotar as barras
plt.bar(x_pos, std_devs, align='center', alpha=0.5)

# Adicionar os rótulos para o eixo x
plt.xticks(x_pos, labels, rotation=90)

# Adicionar um título e rótulos para os eixos
plt.title('Desvio padrão da coluna _Eihop para cada conjunto de dados')
plt.xlabel('Conjunto de dados')
plt.ylabel('Desvio padrão')

# Mostrar o gráfico
plt.show()
```
